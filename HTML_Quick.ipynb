```python
!pip install gradio transformers accelerate torch
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
model_name = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
import gradio as gr

# -------------------------
# PALETTE DATA
# -------------------------
PALETTES = {
    "modern": {
        "Primary": "#667eea",
        "Secondary": "#764ba2",
        "Accent": "#f97316",
        "Background": "#0f172a",
        "Text": "#f1f5f9"
    },
    "ocean": {
        "Primary": "#2193b0",
        "Secondary": "#6dd5ed",
        "Accent": "#00bcd4",
        "Background": "#0c1e2c",
        "Text": "#e0f2fe"
    },
    "forest": {
        "Primary": "#134e4a",
        "Secondary": "#16a34a",
        "Accent": "#22c55e",
        "Background": "#0f1f17",
        "Text": "#dcfce7"
    },
    # NEW PALETTES ADDED
    "vibrant": {
        "Primary": "#ff416c",
        "Secondary": "#ff4b2b",
        "Accent": "#ffe600",
        "Background": "#1a1a2e",
        "Text": "#ffffff"
    },
    "sunset": {
        "Primary": "#ff7e5f",
        "Secondary": "#feb47b",
        "Accent": "#ffcc70",
        "Background": "#2c1b18",
        "Text": "#fff3e6"
    },
    "luxury": {
        "Primary": "#b993d6",
        "Secondary": "#8ca6db",
        "Accent": "#d4af37",
        "Background": "#0d0d0d",
        "Text": "#f5f5f5"
    },
    "creative": {
        "Primary": "#00c6ff",
        "Secondary": "#0072ff",
        "Accent": "#ff00cc",
        "Background": "#141e30",
        "Text": "#e0e0ff"
    }
}

# -------------------------
# HTML GENERATOR
# -------------------------
def generate_page(title, description, style):
    palette = PALETTES.get(style, PALETTES["modern"])

    return f"""
<!DOCTYPE html>
<html>
<head>
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<style>
:root {{
    --primary:{palette["Primary"]};
    --secondary:{palette["Secondary"]};
    --accent:{palette["Accent"]};
    --bg:{palette["Background"]};
    --text:{palette["Text"]};
}}

body {{
    margin:0;
    font-family:Arial, sans-serif;
    background:var(--bg);
    color:var(--text);
    scroll-behavior: smooth;
}}

.navbar {{
    background:#0b1120;
    padding:20px 60px;
    display:flex;
    justify-content:space-between;
}}

.navbar a {{
    color:var(--text);
    text-decoration:none;
    margin-left:30px;
}}

.hero {{
    height:80vh;
    background:linear-gradient(135deg,var(--primary),var(--secondary));
    display:flex;
    flex-direction:column;
    justify-content:center;
    align-items:center;
    text-align:center;
}}

.hero h1 {{
    font-size:48px;
    margin-bottom:20px;
}}

.hero p {{
    font-size:18px;
}}

.hero button {{
    margin-top:20px;
    padding:14px 35px;
    border:none;
    border-radius:8px;
    background:var(--accent);
    color:white;
    font-weight:bold;
    cursor:pointer;
}}

section {{
    padding:100px 20px;
    text-align:center;
}}

footer {{
    background:#0b1120;
    padding:40px;
    text-align:center;
}}
</style>
</head>

<body>

<div class="navbar">
    <div><strong>{title}</strong></div>
    <div>
        <a href="#features">Features</a>
        <a href="#about">About</a>
        <a href="#contact">Contact</a>
    </div>
</div>

<div class="hero">
    <h1>Welcome to Your Site</h1>
    <p>{description}</p>
    <button onclick="document.getElementById('features').scrollIntoView()">Get Started</button>
</div>

<section id="features">
    <h2>Features</h2>
    <p>Professional layout with auto styling.</p>
</section>

<section id="about">
    <h2>About</h2>
    <p>Clean design, modern UI and responsive structure.</p>
</section>

<section id="contact">
    <h2>Contact</h2>
    <p>Reach out and build something amazing.</p>
</section>

<footer>
    © 2026 {title}. All Rights Reserved.
</footer>

</body>
</html>
"""

# -------------------------
# DARK UI CSS
# -------------------------
dark_css = """
body { background:#0b1120 !important; }

.gradio-container {
    background:#0b1120 !important;
    max-width:1400px !important;
    margin:auto !important;
}

h1, h2, h3, p, label {
    color:#f1f5f9 !important;
}

textarea, input, select {
    background:#111827 !important;
    color:white !important;
    border:1px solid #2d3748 !important;
    border-radius:8px !important;
}

button {
    background:#f97316 !important;
    color:white !important;
    border:none !important;
    border-radius:8px !important;
    font-weight:600 !important;
    padding:12px 20px !important;
    cursor:pointer;
}

button:hover {
    opacity:0.9;
}
"""

# -------------------------
# UI
# -------------------------
with gr.Blocks(css=dark_css) as demo:

    gr.Markdown("""
# 🎨 HTML Quick-Styler Advanced
### Professional HTML Page Generator with Auto-Styling
Generate production-ready HTML pages in seconds.
""")

    with gr.Tabs():

        # Generate Page
        with gr.Tab("📝 Generate Page"):
            with gr.Row():
                with gr.Column():
                    title = gr.Textbox(label="Page Title",
                                       value="Modern Portfolio Website")

                    description = gr.Textbox(
                        label="Page Description & Sections",
                        value="Hero section with welcome message, features list, about section, portfolio gallery, testimonials, contact form, and footer"
                    )

                    style = gr.Dropdown(list(PALETTES.keys()),
                                        value="modern",
                                        label="Color Style")

                    generate_btn = gr.Button("🔥 Generate HTML")

                with gr.Column():
                    output = gr.Code(language="html")

            generate_btn.click(generate_page,
                               [title, description, style],
                               output)

        # Live Preview
        with gr.Tab("👀 Live Preview"):
            preview_btn = gr.Button("📄 Load Preview")
            preview = gr.HTML()

            preview_btn.click(generate_page,
                               [title, description, style],
                               preview)

        # Color Palettes
        with gr.Tab("🎨 Color Palettes"):
            style_radio = gr.Radio(list(PALETTES.keys()),
                                   value="modern",
                                   label="Choose Style")

            palette_btn = gr.Button("🎯 Get Palette")
            palette_output = gr.Markdown()

            palette_btn.click(
                lambda s: f"""
Primary: {PALETTES[s]['Primary']}
Secondary: {PALETTES[s]['Secondary']}
Accent: {PALETTES[s]['Accent']}
Background: {PALETTES[s]['Background']}
Text: {PALETTES[s]['Text']}
""",
                style_radio,
                palette_output
            )

        # Templates
        with gr.Tab("📂 Templates"):

            gr.Markdown("## 🚀 Quick Templates")

            with gr.Row():
                with gr.Column():
                    gr.Markdown("### 🎨 Portfolio Website\nHero, Portfolio Gallery, About, Contact, Footer")
                    p_btn = gr.Button("Use Template")

                with gr.Column():
                    gr.Markdown("### 💼 SaaS Landing Page\nHero, Features, Pricing, Testimonials, CTA, Footer")
                    s_btn = gr.Button("Use Template")

                with gr.Column():
                    gr.Markdown("### 🏢 Company Website\nHeader, About, Services, Team, Contact, Footer")
                    c_btn = gr.Button("Use Template")

            p_btn.click(lambda: ("Modern Portfolio Website",
                                 "Hero, Portfolio Gallery, About, Contact, Footer"),
                        outputs=[title, description])

            s_btn.click(lambda: ("SaaS Landing Page",
                                 "Hero, Features, Pricing, Testimonials, CTA, Footer"),
                        outputs=[title, description])

            c_btn.click(lambda: ("Company Website",
                                 "Header, About, Services, Team, Contact, Footer"),
                        outputs=[title, description])

        # Guide
        with gr.Tab("📘 Guide"):

            with gr.Row():

                with gr.Column():
                    gr.Markdown("""
## 📖 HTML Quick-Styler Guide
How it works

### 1️⃣ Enter Details
- Title: Page name
- Description: Section you want
- Color Style: pick a theme

### 2️⃣ Automatic Detection
The tool recognizes these keywords:

- hero → Hero section
- features → Features Grid
- about → About Section
- portfolio → Portfolio
- testimonial → Testimonial
- pricing → Pricing
- contact → Contact
- footer → Footer
""")

                with gr.Column():
                    gr.Markdown("""
🔥 Production-ready HTML output.
""")

demo.launch()
```
